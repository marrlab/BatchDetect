```python
from pathlib import Path
from tqdm import tqdm
import pandas as pd
```

### Patch naming: \<PATIENT>_\<x>_\<y>.jpeg

```python
base_dir = Path('/lustre/groups/shared/users/peng_marr/BatchDetect/BatchDetectCRC')
```

```python
# rename TCGA patches
patch_list = list(base_dir.glob('**/TCGA*/*.jpeg'))
for p in tqdm(patch_list):
    if 'TCGA' not in p.name:
        new_name = p.parent / f'{p.parent.name[:12]}_{p.name}'
        p.rename(new_name)
```

```python
patch_list = list(base_dir.glob('**/TCGA*/*.jpeg'))
```

```python
# rename CPTAC patches
patch_list = list(base_dir.glob('CPTAC/*/*.jpeg'))
for p in tqdm(patch_list):
    new_name = p.parent / f'{p.parent.name.split("-")[0]}_{p.name.split("_")[-2]}_{p.name.split("_")[-1]}'
    p.rename(new_name)
    # print(new_name)
    # break
```

```python
# rename PAIP patches
patch_list = list(base_dir.glob('PAIP/*/*.jpeg'))
for p in tqdm(patch_list):
    new_name = p.parent / f'training-data-{p.parent.name.split("_")[2]}_{p.name.split("_")[-2]}_{p.name.split("_")[-1]}'
    p.rename(new_name)
    # print(new_name)
    # print(p)
    # break
```
